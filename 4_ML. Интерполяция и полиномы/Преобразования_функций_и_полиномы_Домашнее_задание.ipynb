### №1 $(-2, 2), (-1, 1), (0, 0), (1, 1), (2, 2)$<br><br>

```python
# 5 точек, значит полином 4 степени
a4, a3, a2, a1, a0 = symbols ('a4, a3, a2, a1, a0')
# сводободное слагаемое а0 = 0, т.к. есть точка (0, 0)
```

```python
eq1_lp = a4 * (-2) **4 + a3 * (-2) **3 + a2 * (-2) **2 + a1 * (-2) + a0 - 2
eq2_lp = a4 * (-1) **4 + a3 * (-1) **3 + a2 * (-1) **2 + a1 * (-1) + a0 - 1
eq3_lp = a4 * (0) **4 + a3 * (0) **3 + a2 * (0) **2 + a1 * (0) + a0 - 0
eq4_lp = a4 * (1) **4 + a3 * (1) **3 + a2 * (1) **2 + a1 * (1) + a0 - 1
eq5_lp = a4 * (2) **4 + a3 * (2) **3 + a2 * (2) **2 + a1 * (2) + a0 - 2
```

```python
nonlinsolve([eq1_lp, eq2_lp, eq3_lp, eq4_lp, eq5_lp], [a4, a3, a2, a1, a0])
```

```python
x = Symbol('x')
f = -1/6*x**4 + 7/6*x**2 
```

```python
plot(f, (x,-2,2))
```

```python
print(f.subs(x, -2))
print(f.subs(x, -1))
print(f.subs(x, 0))
print(f.subs(x, 1))
print(f.subs(x, 2))
```

```python
x_r =np.linspace(-3,3, num = 100)
f = lambda x: -1/6*x**4 + 7/6*x**2
```

```python
plt.figure(figsize=(10,5))
plt.axhline(0,color = 'b')
plt.axvline(0,color = 'b')
plt.grid()
plt.plot(x_r,f(x_r))
plt.scatter([-2,-1,0,1,2],[2,1,0,1,2], color = 'g')
plt.show()
```

```python
x = Symbol('x')
f = -1/6*x**4 + 7/6*x**2 
```

```python
# исследование функции
if f.subs(x, 100) == f.subs(x, -100):
    print('четность: функция четная')
else:
    print('четность: функция нечетная') 
print('нули функции:',solve(f))
print('f>0 на:',solve_univariate_inequality (f>0,x))
print('f<0 на:',solve_univariate_inequality (f<0,x))
print('f возрастает на:',solve_univariate_inequality (diff(f)>0,x))
print('f убывает на:',solve_univariate_inequality (diff(f)<0,x))
print('экстремумы:',solve(diff(f)))
```

### №2 $(-3, 33), (-2, 31), (-1, 18), (1, -18), (2, -31), (3, -33)$

```python
# 6 точек, значит полином 5 степени
a5, a4, a3, a2, a1, a0 = symbols ('a5 a4, a3, a2, a1, a0')
```

```python
eq1_lp = a5 * (-3) **5 + a4 * (-3) **4 + a3 * (-3) **3 + a2 * (-3) **2 + a1 * (-3) + a0 - 33
eq2_lp = a5 * (-2) **5 + a4 * (-2) **4 + a3 * (-2) **3 + a2 * (-2) **2 + a1 * (-2) + a0 - 31
eq3_lp = a5 * (-1) **5 + a4 * (-1) **4 + a3 * (-1) **3 + a2 * (-1) **2 + a1 * (-1) + a0 - 18
eq4_lp = a5 * (1) **5 + a4 * (1) **4 + a3 * (1) **3 + a2 * (1) **2 + a1 * (1) + a0 + 18
eq5_lp = a5 * (2) **5 + a4 * (2) **4 + a3 * (2) **3 + a2 * (2) **2 + a1 * (2) + a0 + 31
eq6_lp = a5 * (3) **5 + a4 * (3) **4 + a3 * (3) **3 + a2 * (3) **2 + a1 * (3) + a0 + 33
```

```python
nonlinsolve([eq1_lp, eq2_lp, eq3_lp, eq4_lp, eq5_lp, eq6_lp], [a5, a4, a3, a2, a1, a0])
```
